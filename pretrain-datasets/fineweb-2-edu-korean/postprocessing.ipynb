```python
from datasets import load_from_disk

fw2_edu_kr = load_from_disk("../../artifacts/dataset/fineweb-2-edu-korean-scored")
fw2_edu_kr
```

```python
import os

num_processors = max(1, os.cpu_count() - 8)
print(f"Total CPUs: {os.cpu_count()}, Using {num_processors} processes for mapping.")

fw2_edu_kr_score_gte_3 = fw2_edu_kr.filter(lambda x: x["score"] >= 3, num_proc=num_processors)
print(f"Number of samples with score >= 3: {len(fw2_edu_kr_score_gte_3)}")
print(fw2_edu_kr_score_gte_3[44])

fw2_edu_kr_score_gte_3.push_to_hub("minpeter/fineweb-2-edu-korean", private=False, num_proc=num_processors)
```
